#### 考虑 XXL 规格的模型，仅考虑 FFN 的两个线性层，$d_{model}=16384,d_{ff}=53248,num\_blocks=126$

每个 FFN 块含有两个权重矩阵 $W_1:d_{model}\times d_{ff},W_2:d_{ff}\times d_{model}$,如此参数量是
$$
P_{block}=2\times d_{model}\times d_{ff},P_{total}=num\_blocks\times 2\times d_{model}\times d_{ff}
$$
再代入一下数值
$$
P_{block}=1744830464=1.745\times10^9,P_{total}=126\times 1744830464=2.198\times 10^{11}
$$

真懒得算了QAQ
